## Import Data

```r
hotel_data = read.csv("hotel_booking.csv")
# data source - https://www.kaggle.com/datasets/mojtaba142/hotel-booking?resource=download
```

## Delete Columns with Fake Data

```r
head(hotel_data)
hotel_data <- hotel_data[,!names(hotel_data) %in% c("name", "email", "phone.number", "credit_card")]
str(hotel_data)
```

## Change Variable Type

```r
hotel_data[is.na(hotel_data)] <- 0
hotel_data$hotel <- as.factor(hotel_data$hotel)
# hotel_data$is_canceled <- as.factor(hotel_data$is_canceled)
hotel_data$arrival_date_year <- as.factor(hotel_data$arrival_date_year)
hotel_data$arrival_date_month <- as.factor(hotel_data$arrival_date_month)
hotel_data$arrival_date_week_number <- as.factor(hotel_data$arrival_date_week_number)
hotel_data$arrival_date_day_of_month <- as.factor(hotel_data$arrival_date_day_of_month)
hotel_data$meal <- as.factor(hotel_data$meal)
hotel_data$country <- as.factor(hotel_data$country)
hotel_data$market_segment <- as.factor(hotel_data$market_segment)
hotel_data$distribution_channel <- as.factor(hotel_data$distribution_channel)
hotel_data$is_repeated_guest <- as.factor(hotel_data$is_repeated_guest)
hotel_data$reserved_room_type <- as.factor(hotel_data$reserved_room_type)
hotel_data$assigned_room_type <- as.factor(hotel_data$assigned_room_type)
hotel_data$deposit_type <- as.factor(hotel_data$deposit_type)
hotel_data$agent <- as.factor(hotel_data$agent)
hotel_data$company <- as.factor(hotel_data$company)
hotel_data$customer_type <- as.factor(hotel_data$customer_type)
hotel_data$reservation_status <- as.factor(hotel_data$reservation_status)
hotel_data$deposit_type <- as.factor(hotel_data$deposit_type)
hotel_data$reservation_status_date <- as.factor(hotel_data$reservation_status_date)
str(hotel_data)
```

```r
hotel_data[, c('hotel','arrival_date_year','arrival_date_month','arrival_date_week_number',
               'arrival_date_day_of_month','meal','country','market_segment','distribution_channel',
               'is_repeated_guest','reserved_room_type','assigned_room_type','deposit_type','agent','company',
               'customer_type','reservation_status','reservation_status_date')]<- sapply(hotel_data[,
                c('hotel','arrival_date_year','arrival_date_month','arrival_date_week_number',
               'arrival_date_day_of_month','meal','country','market_segment','distribution_channel',
               'is_repeated_guest','reserved_room_type','assigned_room_type','deposit_type','agent','company',
               'customer_type','reservation_status','reservation_status_date')], unclass)
head(hotel_data)
```

## Create Plots with respect to Binary Response Variable (is_canceled)

```r
library(dplyr)
library(tidyr)
options(repr.plot.width=25, repr.plot.height=15)
mtlong = gather(hotel_data, key = "var", value = "value", -is_canceled) %>%
    group_by(var)
    
ggplot(mtlong, aes(x = value, y = is_canceled)) +
    geom_point(pch = 20, size =3 ) +
    labs(x = "") +
    facet_wrap(~ var, scales = "free")+  geom_count()
```

## Create Plots with respect to Continuous Response Variable (lead_time)

```r
library(dplyr)
library(tidyr)
options(repr.plot.width=25, repr.plot.height=15)
mtlong = gather(hotel_data, key = "var", value = "value", -lead_time) %>%
    group_by(var)
    
ggplot(mtlong, aes(x = value, y = lead_time)) +
    geom_point(pch = 20, size =3 ) +
    labs(x = "") +
    facet_wrap(~ var, scales = "free")+  geom_count()
```

## Calculate Correlation with respect to Binary Response Variable (is_canceled)

```r
corrb = data.frame(matrix(NA, nrow = 32, ncol = 2))
corrb[,1] = colnames(hotel_data)
for (i in 1:32){
   corrb[i,2] =  cor(hotel_data[,2],hotel_data[,i])
}
corrb
corrb_sort <-corrb[order(corrb$X2),]
corrb_sort
```
